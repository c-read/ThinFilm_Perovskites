```python
#! pip install xlrd
import sys
#!{sys.executable} -m pip install scipy
import scipy
import scipy.stats as st
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
import pylab
import numpy as np
from lmfit import minimize, Parameters, Parameter, report_fit

pylab.rcParams['figure.figsize'] = (15.0, 10.0)
n=1
m=0
x=200
hc=6.63*3*(10**-26)
e=1.6*(10**-19)

df = pd.read_excel('190404_450.xlsx', unpack=True, usecols=range(0,30),sheet_name=1)

w=df['PL']
E=hc/(w*e*(10**-9))



T4=df['T4']
T20=df['T20']
T30=df['T30']
T50=(df['T50'])
T70=(df['T70'])
T80=(df['T80'])
T90=(df['T90'])
T100=(df['T100'])
T110=(df['T110'])
T130=(df['T130'])
T150=(df['T150'])
T170=(df['T170'])*1/2
T180=(df['T180'])*1/2
T190=(df['T190'])*1/2
T200=(df['T200'])*1/2
T210=(df['T210'])*1/2
T220=(df['T220'])*1/6
T230=(df['T230'])*1/6
T240=(df['T240'])*1/6
T250=(df['T250'])*1/6
T260=(df['T260'])*1/6
T270=(df['T270'])*1/6
T280=(df['T280'])*1/6
T290=(df['T290'])*1/6
T300=(df['T300'])*1/6
T305=(df['T305'])*1/6


counts = [T305,T300,T290,T280,T270,T260,T250,T240,T230,T220,
          T210,T200,T190,T180,T170,T150,T130,T110,T100,T90,T80,T70,T50,T30,T20,T4]
temps = np.array([305,300,290,280,270,260,250,240,230,220,
          210,200,190,180,170,160,150,130,110,100,90,70,50,30,20,4])


#counts = [T305,T300,T290,T280,T270,T260,T240,T230,T220,
 #         T200,T180,T170,T150,T130,T110,T100,T90,T80,T70,T50,T30,T4]
#temps = np.array([305,300,290,280,270,260,240,230,220,200,180,170,150,130,110,100,90,80,70,50,30,4])




intensity=[]
fwhm=[]
wavelength=[]
raw_w=[]
fwhm_err=[]
peak_err=[]
vweight=[]

for i in counts:
    
    #plt.plot(w,i,label='raw')
    i=savgol_filter(i,61,3)
    plt.plot(w,i,label='smoothed')
    
    x_points=[]
    y_points=[]
    peakw=[]
    
    #identify peak intensity and corresponding wavelength
    for j,k in zip(i,w):
        #make list of PL curve to fit gaussian
        #### you need to adjust this k range depending on how noisy the high temp data is ###
        if  840>k>740:
            x_points.append(k)
            y_points.append(j)
    
    for j,k in zip(y_points,x_points):
        if j==max(y_points):
            peakw.append(k)
            raw_w.append(k)
            intensity.append(j)
    #turn lists to arrays
    peakw=np.array(peakw)
    x_points=np.array(x_points)
    y_points=np.array(y_points)
    
    sigma=np.sqrt(np.sum(x_points-peakw)**2/len(x_points))
    sigma=np.array(sigma)
    print('\n',
          '--- PEAK DATA ---','\n',
          'peakw =',peakw)
    

    def voigt(x_points,A,peakw,sigma,a,B):
        sig = 2*sigma*1/(1+np.exp(a*(x_points-peakw)))
        return (A/sig)*np.sqrt(4*np.log(2)/np.pi)*np.exp(-4*np.log(2)*((x_points-peakw)/sig)**2) + (B*2/(sig*np.pi)*(1/(4*((x_points-peakw)/sig)**2+1)))
    popt,pcov=scipy.optimize.curve_fit(voigt,x_points,y_points,p0=[1,peakw,sigma,0,1],maxfev=10000)
    fwhm_err.append(np.sqrt(pcov[2,2]))
    peak_err.append(np.sqrt(pcov[1,1]))
    print('error in fwhm =',np.sqrt(pcov[2,2]))
    print('error in peak wavelength=',np.sqrt(pcov[1,1]))

    #determine weighting f from A and B values
    f=popt[0]/(popt[0]+popt[4])
    print('\n',
          ' --- VOIGT ---','\n',
          'A =', popt[0],'\n',
          'B =',popt[4],'\n',
           'peak =',popt[1],'\n',
          'FWHM =',popt[2],'\n',
          'a =',popt[3],'\n',
          'f =',f)
    print('popt',*popt)
    plt.plot(x_points,voigt(x_points,*popt),label='voigt')
    fwhm.append(popt[2])
    wavelength.append(popt[1])
    vweight.append(f)
    
    def fit_fc(params, x_points, y_points):

        a = params['a'].value
        A = params['A'].value
        peakw = params['peakw'].value
        sigma = params['sigma'].value
        f = params['f'].value
    
        #introduce asymmetry component
        sig = 2*sigma*1/(1+np.exp(a*(x_points-peakw)))
        model = (f*A/sig)*np.sqrt(4*np.log(2)/np.pi)*np.exp(-4*np.log(2)*((x_points-peakw)/sig)**2) + ((1-f)*A*2/(sig*np.pi)*(1/(4*((x_points-peakw)/sig)**2+1)))
        return model-y_points #that's what you want to minimize

    # create a set of Parameters
    # 'value' is the initial condition
    # 'min' and 'max' define your boundaries
    params = Parameters()
    params.add('a',value=0)
    params.add('A', value= 1000, min=0)
    params.add('peakw', value= peakw, min=750, max=900)
    params.add('sigma', value= sigma)
    params.add('f', value= 0.5, min=0.0, max=1.0)

    # do fit, here with leastsq model
    #result = minimize(fit_fc, params, args=(x_points, y_points))
    #write error report
    #report_fit(result.params)
    #print(result.params)
    #plt.plot(x_points,fit_fc(result.params,x_points,y_points)+y_points,label='lmfit')

    plt.legend()
    plt.xlabel('Wavelength/nm',fontsize='15')
    plt.ylabel('Counts',fontsize='15')
    #plt.xlim(700,825)
    
    print('PL at',temps[m],'K')
    m=m+1
    plt.show()
    

#plt.errorbar(temps,gamma,xerr=0,yerr=fwhm_err)
plt.plot(temps,fwhm,'o')
plt.title('0.08% Cs - I2.53/Br0.47',fontsize='15')
plt.xlabel('Temperature / K',fontsize='15')
plt.xticks(fontsize='12')
#plt.ylim(15,60)
#plt.xlim(-10,320)
plt.ylabel('FWHM',fontsize='15')
plt.yticks(fontsize='12')
plt.show()

#normalise intensity
#intensity=(intensity-(min(intensity)))/(max(intensity)-min(intensity))
plt.plot(temps,intensity,'o')
plt.title('0.08% Cs - I2.53/Br0.47',fontsize='15')
plt.xlabel('Temperature / K',fontsize='15')
plt.xticks(fontsize='12')
#plt.ylim(-200,12000)
#plt.xlim(-10,320)
plt.ylabel('Normalised Peak Photoluminescence',fontsize='15')
plt.yticks(fontsize='12')
plt.show()

plt.plot(temps,wavelength)
plt.title('0.08% Cs - I2.53/Br0.47',fontsize='15')
plt.xlabel('Temperature / K',fontsize='15')
plt.xticks(fontsize='12')
#plt.ylim(750,790)
#plt.xlim(-10,320)
plt.ylabel('Peak Wavelength',fontsize='15')
plt.yticks(fontsize='12')
plt.show()

#create array of T,peak wavelength, Peak PL, fwhm
results=np.column_stack((temps,wavelength,intensity,vweight,fwhm,fwhm_err,peak_err,raw_w))
np.savetxt('cs3.txt',results)


#plot intensity/energy
for i in counts:
    x=10**-22
    i=savgol_filter(i,31,3)
    #add jacobian scaling factor as converting wavelength to energy
    i=i*(hc/E**2)
    i = i + n*x
    n = n + 1
    plt.plot(E,i)

plt.legend(fontsize='large',loc='right', labelspacing=-3)
plt.xlabel('Energy/eV')
plt.ylabel('counts')
plt.xlim(1.4,1.8)
plt.show()
```
